```r
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```r
df = pd.read_csv("/kaggle/input/customer/CustomerChurn.csv")
df.head(7044)
```

```r
df.info()
df.describe()
```

```r
print(df.isnull().sum())
```

```r
plt.figure(figsize=(8,5))
sns.histplot(df['MonthlyCharges'], kde=True, bins=50)
plt.title("Distribution of Monthly Charges")
plt.show()
```

```r
plt.figure(figsize=(6,4))
sns.countplot(x='Churn', data=df, palette='coolwarm')
plt.title("Churn Distribution")
plt.show()
```

```r
plt.figure(figsize=(6,4))
sns.countplot(x='gender', data=df, palette='coolwarm')
plt.title("Gender Distribution")
plt.show()
```

```r
plt.figure(figsize=(8,5))
sns.countplot(x='Contract', hue='Churn', data=df, palette='coolwarm')
plt.title("Churn Rate by Contract Type")
plt.show()
```

```r
plt.figure(figsize=(8,5))
sns.histplot(df["tenure"], bins=30, kde=True, color="purple")
plt.title("Distribution of Tenure")
plt.xlabel("Tenure (Months)")
plt.ylabel("Count")
plt.show()
```

```r
plt.figure(figsize=(8,5))
sns.scatterplot(x="MonthlyCharges", y="TotalCharges", hue="Churn", data=df, alpha=0.7)
plt.title("Monthly Charges vs. Total Charges")
plt.xlabel("Monthly Charges ($)")
plt.ylabel("Total Charges ($)")
plt.legend(title="Churn", loc="upper left")
plt.show()
```

```r
plt.figure(figsize=(8,5))
sns.countplot(x="Contract", hue="Churn", data=df, palette="Set2")
plt.title("Churn Rate by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Count")
plt.legend(title="Churn")
plt.show()
```

```r
plt.figure(figsize=(8,5))
sns.countplot(x="InternetService", hue="Churn", data=df, palette="husl")
plt.title("Churn by Internet Service Type")
plt.xlabel("Internet Service")
plt.ylabel("Count")
plt.legend(title="Churn")
plt.show()
```

```r
# Install necessary packages if not already installed
install.packages(c("ggplot2", "dplyr", "readr"))

# Load libraries
library(ggplot2)
library(dplyr)
library(readr)

# Read the dataset
df <- read_csv("/kaggle/input/customer/CustomerChurn.csv")

# View first few rows
head(df)

# Check structure of the dataset
str(df)

# Summary statistics
summary(df)
```

```r
ggplot(df, aes(x = Churn, fill = Churn)) +
  geom_bar() +
  theme_minimal() +
  ggtitle("Churn Distribution") +
  xlab("Churn") +
  ylab("Count") +
  scale_fill_manual(values = c("#FF5733", "#33FF57"))
```

```r
ggplot(df, aes(x = tenure)) +
  geom_histogram(bins = 30, fill = "blue", color = "black", alpha = 0.7) +
  theme_minimal() +
  ggtitle("Distribution of Tenure") +
  xlab("Tenure (Months)") +
  ylab("Count")
```

```r
ggplot(df, aes(x = MonthlyCharges, y = TotalCharges, color = Churn)) +
  geom_point(alpha = 0.6) +
  theme_minimal() +
  ggtitle("Monthly Charges vs. Total Charges") +
  xlab("Monthly Charges ($)") +
  ylab("Total Charges ($)") +
  scale_color_manual(values = c("#E74C3C", "#2ECC71"))
```

```r
ggplot(df, aes(x = Contract, fill = Churn)) +
  geom_bar(position = "dodge") +
  theme_minimal() +
  ggtitle("Churn by Contract Type") +
  xlab("Contract Type") +
  ylab("Count") +
  scale_fill_manual(values = c("#3498DB", "#E74C3C"))
```

```r
ggplot(df, aes(x = InternetService, fill = Churn)) +
  geom_bar(position = "dodge") +
  theme_minimal() +
  ggtitle("Churn by Internet Service") +
  xlab("Internet Service") +
  ylab("Count") +
  scale_fill_manual(values = c("#F39C12", "#16A085"))
```
